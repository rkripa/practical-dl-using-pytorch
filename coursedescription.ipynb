## Course Descripon

Deep Learning (DL) is a subfield of Machine Learning (ML) which again is a subfield of Artificial Intelligence (AI).  DL uses algorithms which are inspired by how our brain is structured and functions.  In DL you build models which are called Networks.  The models solve a lot of practical problems such as image classification, object detection and language translation etc.  They can predict home and stock prices, face recognition, classify cancer as malign or benign and more.

In this course, students use PyTorch which is an open-source deep learning research platform that provides maximum flexibility and speed.  PyTorch was developed by Facebook Research.  PyTorch supports computation on both CPUs and GPUs. It is very popular among researchers.

Students will learn/build models of progressively increasing complexity using industry standard datasets such as mnist, imdb, Boston housing prices, iris, imagenet, glove etc.

By the end of the class, students will have basic understanding of DL and training techniques,  an environment to experiment and explore DL projects. The class prepares students to pursue a career as a DL Engineer and set the foundation for a DL Researcher.


### Topics Include:
